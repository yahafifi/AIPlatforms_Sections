```python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
pipeline = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2-1-base").to(device)
```

```python
prompt = input('Enter prompt: ')
image = pipeline(prompt).images[0]
```

```python
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
import torch

from diffusers import StableVideoDiffusionPipeline
from diffusers.utils import load_image, export_to_video

pipe = StableVideoDiffusionPipeline.from_pretrained(
  "stabilityai/stable-video-diffusion-img2vid-xt", torch_dtype=torch.float16, variant="fp16"
)
pipe.enable_model_cpu_offload()

# Load the conditioning image
image = load_image("https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/diffusers/svd/rocket.png")
image = image.resize((1024, 576))
```

```python
generator = torch.manual_seed(42)
frames = pipe(image, decode_chunk_size=8, generator=generator).frames[0]

export_to_video(frames, "generated.mp4", fps=7)
```

```python
summarizer = pipeline("summarization")
```

```python
text = """
Artificial Intelligence (AI) is transforming industries worldwide. It refers to 
the simulation of human intelligence in machines programmed to think and learn 
like humans. AI applications include natural language processing, robotics, and 
computer vision, which enable tasks such as speech recognition, facial detection, 
and autonomous vehicle control. As technology advances, AI continues to play a 
significant role in various fields, including healthcare, finance, and education.
"""
```

```python
summary = summarizer(text, max_length=60, min_length=20, do_sample=False)
```

```python
print("Summary:", summary[0]['summary_text'])
```
